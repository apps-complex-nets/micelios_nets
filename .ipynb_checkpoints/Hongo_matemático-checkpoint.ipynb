```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, box, Point
from scipy.spatial import KDTree
import networkx as nx
import random
```

```python
# ============================================================
#  BOUNDING BOX DEL DOMINIO

def bounding_box(domain_poly):
    minx, miny, maxx, maxy = domain_poly.bounds
    return box(minx, miny, maxx, maxy)

# ============================================================
# SUBDIVIDIR EN n×n SUBCELDAS

def subdividir(celda, n=3):
    minx, miny, maxx, maxy = celda.bounds
    w = (maxx - minx) / n
    h = (maxy - miny) / n

    subceldas = []
    for i in range(n):
        for j in range(n):
            sub = box(minx + i*w, miny + j*h,
                      minx + (i+1)*w, miny + (j+1)*h)
            subceldas.append(sub)
    return subceldas

# ============================================================
#  SELECCIONAR k CELDAS (DANDO PREFERENCIA A VECINAS)

def seleccionar_k(subceldas, k):
    # Primera celda aleatoria
    seleccion = [random.choice(subceldas)]

    def es_vecina(a, b):
        return a.touches(b) or a.intersects(b)

    while len(seleccion) < k:
        restantes = [c for c in subceldas if c not in seleccion]
        vecinas = [c for c in restantes if any(es_vecina(c, s) for s in seleccion)]

        if vecinas:
            seleccion.append(random.choice(vecinas))
        else:
            seleccion.append(random.choice(restantes))

    return seleccion

# ============================================================
#  GENERAR PUNTOS EN UNA CELDA

def puntos_en_celda(celda, m=3):
    minx, miny, maxx, maxy = celda.bounds
    pts = []
    for _ in range(m):
        x = np.random.uniform(minx, maxx)
        y = np.random.uniform(miny, maxy)
        pts.append((x, y))
    return pts

# ============================================================
#  GENERAR PUNTOS FRACTALES

def generar_puntos_fractales(domain_poly, Df=1.48, n=3, niveles=4, m_por_celda=3):

    k = int(round(n ** Df))             # número de subceldas a conservar
    print("k =", k)

    BB = bounding_box(domain_poly)
    celdas_activas = [BB]

    for nivel in range(niveles):

        nuevas = []

        for C in celdas_activas:

            subceldas = subdividir(C, n)
            seleccionadas = seleccionar_k(subceldas, k)

            for S in seleccionadas:
                inter = S.intersection(domain_poly)
                if not inter.is_empty:
                    nuevas.append(inter)

        celdas_activas = nuevas
        print(f"Nivel {nivel+1}: {len(celdas_activas)} celdas activas")

  # ============================================================
  # Generar puntos dentro de las celdas activas

    puntos = []
    for C in celdas_activas:
        puntos.extend(puntos_en_celda(C, m=m_por_celda))

    return puntos, celdas_activas

```

```python
# CONSTRUIR ÁRBOL GENERADOR MÍNIMO (MST)

def construir_arbol_minimo(puntos):
    """
    Construye un árbol generador mínimo (MST).
    """

    n = len(puntos)
    G = nx.Graph()

    # Añadir vértices
    for i, (x,y) in enumerate(puntos):
        G.add_node(i, pos=(x, y))

    # Matriz de distancias
    dist = np.zeros((n,n))
    for i in range(n):
        xi, yi = puntos[i]
        for j in range(i+1, n):
            xj, yj = puntos[j]
            d = np.hypot(xi - xj, yi - yj)
            dist[i,j] = dist[j,i] = d

    # Grafica completa auxiliar
    H = nx.Graph()
    for i in range(n):
        for j in range(i+1, n):
            H.add_edge(i, j, weight=dist[i,j])

    # MST
    mst = nx.minimum_spanning_tree(H)

    for u, v, data in mst.edges(data=True):
        G.add_edge(u, v, weight=data["weight"])

    return G
```

```python
def genera_arbol(Df, n, niveles, m_por_celda):
    # una elipse (simula un casco convexo real)
    theta = np.linspace(0, 2*np.pi, 200)
    x = 20*np.cos(theta)
    y = 10*np.sin(theta)
    dominio = Polygon(np.column_stack([x, y]))
    
    # 1. Generar puntos fractales dentro del dominio real
    # Df - dimensión fractal
    # nxn rejillas (iniciales) para la dimensión fractal
    # niveles numero de veces que se incrementa la rejilla
    # m_por_celda - punto en cada cuadrito
    puntos, celdas = generar_puntos_fractales(
        domain_poly = dominio,
        Df = Df,
        n = n,
        niveles = niveles,
        m_por_celda = m_por_celda
    )
    
    # 2. Construir el árbol
    G = construir_arbol_minimo(puntos)
    
    # 3. Dibujar
    plt.figure(figsize=(7,7))
    pos = nx.get_node_attributes(G, "pos")
    
    nx.draw(G, pos, node_size=2, edge_color="gray", width=0.6)
    xD, yD = dominio.exterior.xy
    plt.plot(xD, yD, color="blue", linewidth=2.5)
    
    plt.title("Hongo matemático")
    plt.axis("equal")
    plt.show()
```

```python
# Df - dimensión fractal
# nxn rejillas (iniciales) para la dimensión fractal
# niveles numero de veces que se incrementa la rejilla
# m_por_celda - punto en cada cuadrito
genera_arbol(Df=1.48, n=3, niveles=2, m_por_celda=2)
```

```python
import numpy as np
import networkx as nx

def construir_arbol_minimo_restringido(puntos, L_max):
    """
    Construye un árbol generador mínimo donde
    solo se permiten aristas con longitud <= L_max.

    Si el arbol resultante no es conexo, el MST no existe
    y se devuelve None.
    """
    n = len(puntos)
    G = nx.Graph()

    # Añadir nodos con posición
    for i, (x, y) in enumerate(puntos):
        G.add_node(i, pos=(x, y))

    # Matriz de distancias
    dist = np.zeros((n, n))
    for i in range(n):
        xi, yi = puntos[i]
        for j in range(i+1, n):
            xj, yj = puntos[j]
            d = np.hypot(xi - xj, yi - yj)
            dist[i, j] = dist[j, i] = d

    # Grafo auxiliar con SOLO aristas de longitud <= L_max
    H = nx.Graph()
    for i in range(n):
        for j in range(i+1, n):
            d = dist[i, j]
            if d <= L_max:
                H.add_edge(i, j, weight=d)

    # Verificar si H es conexo
    if not nx.is_connected(H):
        print("⚠️ Con L_max =", L_max, "el arbol es disconexo")
        print("   Puedes aumentar L_max o trabajar con componentes por separado.")
        return None

    # MST sobre el grafo restringido
    mst = nx.minimum_spanning_tree(H)

    # Copiar aristas al grafo final
    for u, v, data in mst.edges(data=True):
        G.add_edge(u, v, weight=data["weight"])

    return G
```

```python
# Df - dimensión fractal
# nxn rejillas (iniciales) para la dimensión fractal
# niveles numero de veces que se incrementa la rejilla
# m_por_celda - punto en cada cuadrito

def arbol_restringido(Df, n, niveles, m_por_celda):
    puntos, celdas = generar_puntos_fractales(
            domain_poly = dominio,
            Df = Df,
            n = n,
            niveles = niveles,
            m_por_celda = m_por_celda
        )
    L_max = 5  # por ejemplo, en las mismas unidades que tus coordenadas
    G_restringido = construir_arbol_minimo_restringido(puntos, L_max)
    
    if G_restringido is not None:
        pos = nx.get_node_attributes(G_restringido, "pos")
        plt.figure(figsize=(7,7))
        nx.draw(G_restringido, pos, node_size=2.0, edge_color="gray", width=0.7)
        plt.axis("equal")
        plt.show()
```
