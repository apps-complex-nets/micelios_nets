```python
import networkx as nx
import os
import pandas as pd
```

```python
PATH_TRICHODERMA = "data/Trichoderma_atroviride/Muestra"
PATH_NEUROSPORA = "data/Neurospora_crassa/Muestra"
PATH_NETS_TRICHODERMA = "data/Trichoderma_atroviride/Redes"
PATH_NETS_NEUROSPORA = "data/Neurospora_crassa/Redes"
NUM_MUESTRAS = 5
```

```python
def create_network(nodes1, nodes2):
    edges = [(n1,n2) for n1, n2 in zip(nodes1, nodes2)]
    #print(edges)
    G = nx.Graph()
    G.add_edges_from(edges)
    for node in list(G):
        G.nodes[node]["x"] = node[0]/100
        G.nodes[node]["y"] = node[1]/100
    #print(G.nodes.data())
    return G
```

```python
def read_coordinates(file_net):
    df = pd.read_csv(file_net, names=["x1", "y1", "x2", "y2"], header=None)
    #Nodos 1
    x1 = df["x1"].tolist()
    y1 = df["y1"].tolist()
    nodes1 = [(f, b) for f, b in zip(x1, y1)]
    #Nodos 2
    x2 = df["x2"].tolist()
    y2 = df["y2"].tolist()
    nodes2 = [(f, b) for f, b in zip(x2, y2)]
    print(file_net)
    print(df)
    return nodes1, nodes2
```

```python
for i in range(NUM_MUESTRAS): 
    files_trichoderma = [PATH_TRICHODERMA + str(i+1) + "/" + f for f in os.listdir(PATH_TRICHODERMA+str(i+1)) if '.csv' in f]
    files_neurospora = [PATH_NEUROSPORA + str(i+1) + "/" + f for f in os.listdir(PATH_NEUROSPORA+str(i+1)) if '.csv' in f]
    files_net_trichoderma = [PATH_TRICHODERMA + str(i+1) + "/" + f[:-4] + ".gexf" for f in os.listdir(PATH_TRICHODERMA+str(i+1)) if '.csv' in f]
    files_net_neurospora = [PATH_NEUROSPORA + str(i+1) + "/" + f[:-4] + ".gexf" for f in os.listdir(PATH_NEUROSPORA+str(i+1)) if '.csv' in f]
    for f_coord, f_net in zip(files_trichoderma, files_net_trichoderma):
        print(f_net)
        nodes1, nodes2 = read_coordinates(f_coord)
        G = create_network(nodes1, nodes2)
        #print(G.edges)
        nx.write_gexf(G, f_net)
    for f_coord, f_net in zip(files_neurospora, files_net_neurospora):
        print(f_net)
        nodes1, nodes2 = read_coordinates(f_coord)
        G = create_network(nodes1, nodes2)
        nx.write_gexf(G, f_net)
```
