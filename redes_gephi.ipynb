```python
import networkx as nx
import os
import pandas as pd
import constants as const
```

```python
def create_network(nodes1, nodes2):
    edges = [(n1,n2) for n1, n2 in zip(nodes1, nodes2)]
    #print(edges)
    G = nx.Graph()
    G.add_edges_from(edges)
    for node in list(G):
        G.nodes[node]["x"] = node[0]/100
        G.nodes[node]["y"] = node[1]/100
    #print(G.nodes.data())
    return G
```

```python
def read_coordinates(file_net):
    df = pd.read_csv(file_net, names=["x1", "y1", "x2", "y2"], header=None)
    #Nodos 1
    x1 = df["x1"].tolist()
    y1 = df["y1"].tolist()
    nodes1 = [(f, b) for f, b in zip(x1, y1)]
    #Nodos 2
    x2 = df["x2"].tolist()
    y2 = df["y2"].tolist()
    nodes2 = [(f, b) for f, b in zip(x2, y2)]
    print(file_net)
    print(df)
    return nodes1, nodes2
```
